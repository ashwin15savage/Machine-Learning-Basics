```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth
```

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Hari', 'Eggs', 'Yogurt'],
           ['Dill', 'Ashwin', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Bhavani shankar', 'Ice cream', 'Eggs']]
```

```python
te=TransactionEncoder()#transform the dataset into array format
te_ary=te.fit(dataset).transform(dataset)#converts the dataset to a boolean array
```

```python
te_ary
```

```python
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
frequent_itemsets = fpgrowth(df, min_support=0.6, use_colnames=True)
frequent_itemsets
#fpgrowth: used for extracting frequent datasets
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
#want rules rom the frequent itemsets only if the level of confidence is above the 70 percent threshold
```

```python
#interested in rules that have a lift score of >= 1.2
#we can use any metric of interest
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules
```

```python
#for more filteration
rules["antecedent_length"]=rules["antecedents"].apply(lambda x: len(x))
rules
```

```python
rules[ (rules['antecedent_length'] >= 2) &
       (rules['confidence'] > 0.75) &
       (rules['lift'] > 1.2) ]
```

# Market basket analysis

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_excel('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx')
df
```

```python
df['Description']=df['Description'].str.strip()
df.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
df['InvoiceNo']=df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
basket = (df[df['Country'] =="France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules[ (rules['lift'] >= 6) &
       (rules['confidence'] >= 0.8) ]
```

```python
basket2 = (df[df['Country'] =="Germany"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

basket_sets2 = basket2.applymap(encode_units)
basket_sets2.drop('POSTAGE', inplace=True, axis=1)
frequent_itemsets2 = apriori(basket_sets2, min_support=0.05, use_colnames=True)
rules2 = association_rules(frequent_itemsets2, metric="lift", min_threshold=1)

rules2[ (rules2['lift'] >= 4) &
        (rules2['confidence'] >= 0.5)]
```
